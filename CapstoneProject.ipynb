## Now we put this on a Map

```python
# create map of New York using latitude and longitude values 43.647177	-79.381576
map_toronto = folium.Map(location=[43.647177, -79.381576], zoom_start=2)

#places = folium.map.FeatureGroup() 

# add markers to map
for placeID in vegan_places.index:
    label = '{}'.format(vegan_places['name'][placeID])

    lat_val = vegan_places['Latitude'][placeID]
    lng_val = vegan_places['Longitude'][placeID]
    #print (lat_val)
    
    label = folium.Popup(label, parse_html=True)

    folium.CircleMarker(
        [lat_val, lng_val],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

map_toronto
```

## Now lets check which neihborhoods have at least 5 vegan places within 0.5miles (0.8km)

```python
NumLoc = DataFrame(columns = {'Number of Locations'})
for neighborhoodID in merged_left.index:
    numVeg = 0
    for placeID in vegan_places.index:
        coords_1 = (merged_left['Latitude'][neighborhoodID], merged_left['Longitude'][neighborhoodID])
        coords_2 = (vegan_places['Latitude'][placeID], vegan_places['Longitude'][placeID])
        dist = geopy.distance.vincenty(coords_1, coords_2).km
        #print (dist)
        if dist < 2:
            numVeg += 1
            print("+")
    NumLoc = NumLoc.append({'Number of Locations': numVeg}, ignore_index= True)
    
Neighborhood_Vegan = merged_left
Neighborhood_Vegan['Number of Locations'] = NumLoc
```

```python
Neighborhood_Vegan
```

```python

# add Neighborhood markers to map
for neighborhoodID in Neighborhood_Vegan.index:
    label = '{}'.format(Neighborhood_Vegan['Neighborhood'][neighborhoodID])

    lat_val = Neighborhood_Vegan['Latitude'][neighborhoodID]
    lng_val = Neighborhood_Vegan['Longitude'][neighborhoodID]
    #print (lat_val)
    
    label = folium.Popup(label, parse_html=True)

    if Neighborhood_Vegan['Number of Locations'][neighborhoodID]>4:
        folium.CircleMarker(
            [lat_val, lng_val],
            radius=20,
            popup=label,
            color='red',
            fill=True,
            fill_color='#CC200C',
            fill_opacity=0.3,
            parse_html=False).add_to(map_toronto) 
    elif Neighborhood_Vegan['Number of Locations'][neighborhoodID]>1:
        folium.CircleMarker(
            [lat_val, lng_val],
            radius=20,
            popup=label,
            color='orange',
            fill=True,
            fill_color='#CC6621',
            fill_opacity=0.3,
            parse_html=False).add_to(map_toronto)  

map_toronto
```
